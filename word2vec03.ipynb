```python
body = data.body.apply(lambda x: gensim.utils.simple_preprocess(str(x)))
```

```python
data.body[12]
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2,
    workers = 4,
)
```

```python
model.build_vocab(body, progress_per = 1000)
```

```python
model.epochs
```

```python
model.corpus_count
```

```python
model.train(body, total_examples = model.corpus_count, epochs = model.epochs)
```

```python
model.save("./word2vec03.model")
```
