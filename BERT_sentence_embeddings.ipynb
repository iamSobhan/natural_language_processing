```python
import tensorflow_text as text
import tensorflow_hub as hub
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
preprocessing_model_bert = hub.KerasLayer(preprocess_url)
```

```python
test_text = ["I love Python Programming.", "NLP is one of my favorite subjects."]
text_preprocessing = preprocessing_model_bert(test_text)
text_preprocessing.keys()
```

```python
bert_model = hub.KerasLayer(encoder_url)
```

```python
bert_results = bert_model(text_preprocessing)
```
